## Pregunta raices

```python
from scipy import optimize
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (x**3 + 2*x**2 -5)

x = np.linspace(-4, 4, 100)
y = f(x)

plt.plot(x, y)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.title('f(x) = x^3 + 2x^2 - 5')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

En el grafico se puede ver que solo presenta una raiz

#### Biseccion

```python
raiz, r = optimize.bisect(f, -2, 2, full_output=True)
print(raiz)
print(f"Numero de iteraciones: {r.iterations}")
x1 = np.linspace(-2, 2, 100)
y1 = f(x1)

plt.scatter(raiz, 0, color='red')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.plot(x1, y1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Raiz de f(x)-Metodo biseccion')
plt.grid()
plt.show()
```

#### Secante

```python
raiz2, r2 = optimize.newton(f, 1, full_output=True)
print(raiz2)
print(f"Numero de iteraciones: {r2.iterations}")

x2 = np.linspace(-2, 2, 100)
y2 = f(x2)

plt.scatter(raiz2, 0, color='red')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.plot(x2, y2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Raiz de f(x)-Metodo secante')
plt.grid()
plt.show()
```

```python
raiz3, r3 = optimize.newton(f, 2, full_output=True)
print(f"Numero de iteraciones: {r3.iterations}")
print(raiz3)
x3 = np.linspace(-2, 2, 100)
y3 = f(x3)

plt.scatter(raiz3, 0, color='red')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.plot(x3, y3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Raiz de f(x)-Metodo secante')
plt.grid()
plt.show()
```
