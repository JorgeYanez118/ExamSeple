#### Parte 1

```python
from src import gauss_seidel, gauss_jacobi
import matplotlib.pyplot as plt
```

```python
A = [[1, 1], [-2, 5]]
b = [7, 0]
x0 = [0, 0]
tol = 0.1
iter = 100

x1j_values = []
x2j_values = []

for i in range(iter):
    x1j, x2j = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=i+1)
    x1j_values.append(x1j)
    x2j_values.append(x2j)

plt.plot(x1j_values, x2j_values, 'bo-', label='Trayectoria')
plt.plot(x1j,x2j, 'ro', label='Solucin')
plt.xlabel('x1j')
plt.ylabel('x2j')
plt.title('Gauss-Jacobi trayectoria')
plt.grid(True)
plt.show()

print(x1j)
print(x2j)



```

```python
x1, x2 = gauss_seidel(A=A, b=b, x0=x0, tol=tol, max_iter=iter)

x1_values = []
x2_values = []

for i in range(iter):
    x1, x2 = gauss_seidel(A=A, b=b, x0=x0, tol=tol, max_iter=i+1)
    x1_values.append(x1)
    x2_values.append(x2)

plt.plot(x1_values, x2_values, 'bo-', label='Gauss-Seidel')
plt.plot(x1,x2, 'ro', label='Solucin')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Gauss-Seidel trayectoria')
plt.legend()
plt.grid(True)
plt.show()
print(x1)
print(x2)
```
